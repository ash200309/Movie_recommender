# Preprocessing

```python
movies= movie.merge(credits,on="title")
print(movie.shape)
movies.shape
```

```python
movies.head(2)
```

```python
movies.info()
#cloumns chosen: #genre #id #keywords #title #overview #release date #cast #crew
```

```python
#cloumns chosen: #genre #id #keywords #title #overview #release date #cast #crew
dataset=movies[["id","title","genres","keywords","overview","release_date","cast","crew"]].copy()
dataset.head(2)
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.duplicated().sum()
```

```python
# correcting columns format: genres
# Using a loop to take only the "name" of genres column
# using ast module to convert string into list for the loop
from ast import literal_eval as le
def convert(obj):
    L=[]
    for i in le(obj):
        L.append(i['name'])
    return L
```

```python
dataset['genres']=dataset['genres'].apply(convert)
dataset.head(1)
```

```python
dataset["genres"][0][0]
```

```python
dataset['keywords']=dataset['keywords'].apply(convert)
dataset.head(1)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in le(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break 
    return L
```

```python
# Since only first 3 are usually the major actors
dataset["cast"]=dataset["cast"].apply(convert3)
dataset.head(1)
```

```python
# job = director is more relevant
def search_director(obj):
    L=[]
    for i in le(obj):
        if i['job']=="Director":
            L.append(i['name'])
            break
    return L
```

```python
dataset["crew"]=dataset["crew"].apply(search_director)
dataset.head(5)
```

```python
dataset['overview'][0]
```

```python
# splitting keywords
dataset['overview']=dataset['overview'].apply(lambda x:x.split())
dataset.head()
```

```python
# Removing blank space between words

dataset['genres']=dataset['genres'].apply(lambda x:[i.replace(" ","") for i in x])
dataset['keywords']=dataset['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
dataset['crew']=dataset['crew'].apply(lambda x:[i.replace(" ","") for i in x])
dataset['cast']=dataset['cast'].apply(lambda x:[i.replace(" ","") for i in x])
dataset.head(2)
```

```python
# Joining columns
dataset['tags']=dataset['genres']+dataset['keywords']+dataset['overview']+dataset['cast']+dataset["crew"]
dataset.head(3)
```

```python
df=dataset[['id','title','tags']].copy()
df.head()
```

```python
# Converting tags column from list to string 
df['tags']=df['tags'].apply(lambda x:" ".join(x))
df.head()
```

```python
# Changing text to lower case
df['tags']=df['tags'].apply(lambda x:x.lower())
df.head()
```

```python
# Stemming changing words like action,actions to action
# using from nltk.stem.porter import PorterStemmer
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags']=df['tags'].apply(stem)
df.head()
```

# Vectorization

```python
# max features= no. of words most repeated  stopwords= if ,of,in,a etc.
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
vector=cv.fit_transform(df["tags"]).toarray()
vector
```

```python
# Similarity object is an array of cosine ditances of each movie from every other movie including itslef
from sklearn.metrics.pairwise import cosine_similarity
similarity= cosine_similarity(vector)
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
# top 5 similar movies
```

```python
# We need to create a function that will return top 5 similar movies 
def recommend(movie):
    movie_index=df[df['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Batman')
```

```python
recommend("Pirates of the Caribbean: At World's End")
```
